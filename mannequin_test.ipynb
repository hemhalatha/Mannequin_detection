```python
import cv2
from ultralytics import YOLO

# Load your model from Google Drive if needed
model = YOLO('/content/drive/MyDrive/yolov8_backups/models/last.pt')
print("Model loaded successfully")
```

```python
from google.colab.patches import cv2_imshow

def detect_sleep(image_path):
    # Read image
    frame = cv2.imread(image_path)

    if frame is None:
        print("‚ùå Error: Image not found:", image_path)
        return

    # Resize to 512x512
    frame_resized = cv2.resize(frame, (512, 512))

    # Predict
    results = model(frame_resized)

    # Annotate
    annotated_frame = results[0].plot()

    # Display in colab
    cv2_imshow(annotated_frame)
```
